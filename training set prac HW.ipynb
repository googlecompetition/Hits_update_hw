### Hits: how many products viewed

```python
df_train['hits'][0]

product_count = []
duration = []
transaction = []
promotion = []
#transac_p = []
transac_h= []
page_level_3 = []
pro_price = []

for i in range(0,len(df_train)):
    hit = df_train['hits'].iloc[i]
    c = 0
    d = 0
    t = 0
    p = 0
    #t_p = 0
    t_h = 0
    page_c = []
    p_price = []
    
    for j in range(0,len(hit)):
         c += len(hit[j]['product'])
         d += int(hit[j]['time'])
         t += 1 if 'transaction' in hit[j] and len(hit[j]['transaction']) > 0 else 0
         p += 1 if 'promotion' in hit[j] and len(hit[j]['promotion']) > 0 else 0
         page_c.append(hit[j]['page']['pagePathLevel3'])
         p_p = 0
         for k in range(0, len(hit[j]['product'])):
            p_p += float(hit[j]['product'][k]['productPrice'])
         p_price.append(p_p)

    
    t_h = t/len(hit) 
        
    product_count.append(c)
    duration.append(d)
    transaction.append(t)
    promotion.append(p)
    transac_h.append(t_h)
    page_level_3.append(page_c)
    pro_price.append(p_price)
```

```python
df_train_hits = pd.concat([df_train,pd.DataFrame({
             'product_count':product_count,
             'duration':duration,
             'transaction':transaction,
             'promotion':promotion,
             'transac_h': transac_h,
             'page_level_3': page_level_3,
             'pro_price': pro_price
              })], axis = 1)


df_train_hits['pro_price_total'] = [sum(df_train_hits['pro_price'][i]) for i in range(0,len(df_train_hits))]
df_train_hits.head()
```
